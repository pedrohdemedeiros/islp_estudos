```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
college = pd.read_csv('college.csv')
college2 = pd.read_csv('college.csv', index_col=0)
college3 = college.rename({'Unnamed: 0': 'College'}, axis= 1)
college3 =college3.set_index('College')
```

```python
college = college3
college
```

```python
college.describe().round(2)
```

```python
colunas = ['Top10perc', 'Apps', 'Enroll']
pd.plotting.scatter_matrix(college[colunas], figsize=(12,12), diagonal='hist')
plt.show()
```

```python
college.boxplot(column='Outstate', by='Private')
plt.show()
```

```python
college['Elite'] = pd.cut(college['Top10perc'], [0,0.5,1], labels = ['No', 'Yes'])
college['Elite'].value_counts()
```

```python
college.boxplot(column='Outstate', by='Elite')
plt.show()
```
