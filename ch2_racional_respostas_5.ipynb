### Em regressão e classificação:

- **Flexibilidade**: A flexibilidade é importante, especialmente nas áreas de fronteira. 
    - **Classificação**: Alta flexibilidade pode ser muito suscetível a ruídos que camuflam as principais características que levariam à classificação de um objeto, enquanto baixa flexibilidade pode obscurecer elementos secundários importantes. Alta flexibilidade é favorável para uma melhor definição dos objetos de fronteira, enquanto a baixa flexibilidade foca nos objetos em que queremos ter 100% de certeza de sua classificação.
    - **Regressão**: Com alta flexibilidade, podemos abordar o máximo de objetos com menos certeza acerca do resultado exato. Com baixa flexibilidade, focamos nos objetos mais típicos.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create data
np.random.seed(0)
x = np.linspace(0, 10, 100)
y_linear = 2 * x + 1 + np.random.normal(0, 2, 100)
y_nonlinear = 2 * x + 10 * np.sin(x) + np.random.normal(0, 3, 100)

# Fit linear model
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression().fit(x.reshape(-1, 1), y_linear)
y_pred_linear = lin_reg.predict(x.reshape(-1, 1))

# Fit polynomial model for nonlinear data
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

degree = 5
poly_reg = make_pipeline(PolynomialFeatures(degree), LinearRegression())
poly_reg.fit(x.reshape(-1, 1), y_nonlinear)
y_pred_nonlinear = poly_reg.predict(x.reshape(-1, 1))

# Plotting
fig, axs = plt.subplots(2, 1, figsize=(10, 12))

# Plot for linear data
axs[0].scatter(x, y_linear, color='lightblue', label='Dados reais')
axs[0].plot(x, y_pred_linear, color='red', linewidth=2, label='Modelo Linear')
axs[0].set_title('Dados com Relação Linear e Ajuste Linear')
axs[0].legend()

# Plot for nonlinear data
axs[1].scatter(x, y_nonlinear, color='lightgreen', label='Dados reais')
axs[1].plot(x, y_pred_nonlinear, color='blue', linewidth=2, label=f'Modelo Polinomial (grau {degree})')
axs[1].set_title('Dados com Relação Não Linear e Ajuste Polinomial')
axs[1].legend()

plt.tight_layout()
plt.show()
```
