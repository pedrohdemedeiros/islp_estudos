A idéia é compreender a dinâmica dos erros em um só gráfico.
Além disso foi utilizado o estilo xqcd para parecer um rascunho

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for the model flexibility
x = np.arange(0.0, 10.0, 0.02)

# Squared bias function
def squared_bias(x):
    return .002*(-x+10)**3

# Variance function
def variance(x):
    return .002*x**3 

# Training error function
def training_error(x):
    return 2.38936 - 0.825077*x + 0.176655*x**2 - 0.0182319*x**3 + 0.00067091*x**4

# Bayes error function
def bayes_error(x):
    return x + 1 - x

# Adjusted test error function
def adjusted_test_error(x):
    return squared_bias(x) + variance(x) + bayes_error(x)

# Plotting the graph
plt.figure(figsize=(10, 8))
plt.plot(x, squared_bias(x), label='squared bias')
plt.plot(x, variance(x), label='variance')
plt.plot(x, training_error(x), label='training error')
plt.plot(x, adjusted_test_error(x), label='adjusted test error', linestyle='--', color='purple')
plt.plot(x, bayes_error(x), label='Bayes error')
plt.legend(loc='upper center')
plt.xlabel('model flexibility')
plt.ylabel('Error')
plt.title("Adjusted Graph")
plt.grid(True)
plt.tight_layout()
plt.show()
```
